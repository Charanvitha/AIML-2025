1.Reading the **data** **bold text**

```python
df = pd.read_csv('employee_data.csv')
print("Original Dataset:")
print(df)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Job_Satisfaction'] = df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mean())
```

```python
print("Dataset after handling missing values:")
print(df)
```

3.Using **MinMaxScaler** for age,salary

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
```

```python
print(df)
```

```python
df['Age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
```

```python
df['Salary'] = (df['Salary'] - df['Salary'].min()) / (df['Salary'].max() - df['Salary'].min())
```

```python
print(df)
```

4.Standardization (Z-score normalization) (for Job_Satisfaction, Work_Hours_Per_Week)

```python
from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
```

```python
df[['Job_Satisfaction']] = s_scaler.fit_transform(df[[ 'Job_Satisfaction']])
```

```python
df[['Work_Hours_Per_Week']] = s_scaler.fit_transform(df[[ 'Work_Hours_Per_Week']])
```

```python
df['Job_Satisfaction']
```

```python
df['Work_Hours_Per_Week']
```

Manually Implementing Standard Scaling Z-Score formula:

```python
df['Job_Satisfaction'] = (df['Job_Satisfaction'] - df['Job_Satisfaction'].mean()) / df['Job_Satisfaction'].std()
```

```python
df['Work_Hours_Per_Week'] = (df['Work_Hours_Per_Week'] - df['Work_Hours_Per_Week'].mean()) / df['Work_Hours_Per_Week'].std()
```
